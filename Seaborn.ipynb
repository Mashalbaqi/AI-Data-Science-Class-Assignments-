```python
# What is Seaborn
# Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", data=data)

plt.show()
```

```python

tips = sns.load_dataset("tips")

sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
# lmplot(): Combines a scatter plot with a regression line.

plt.show()

```

```python
sns.histplot(tips['total_bill'], kde=True)
# histplot(): Plots a histogram with an optional KDE (Kernel Density Estimate) curve.

plt.show()
```

```python
# Create a box plot to show tip distribution across days
sns.boxplot(x="day", y="tip", data=tips)

plt.show()
```

```python
import numpy as np

# Generate random data and compute correlation matrix
data = np.random.rand(5, 5)
corr = np.corrcoef(data)

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```
