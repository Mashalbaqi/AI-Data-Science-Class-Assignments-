```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import sklearn as sk
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Linear Regration

```python
df=pd.read_csv('Salary_Data.csv')
df.head()
```

```python
# x=df.iloc[:,0].values
# y=df.iloc[:,-1].values - this only select according to index, you must select columns according to labels such as salary 

x = np.array(df.YearsExperience).reshape(-1, 1) # input or we can call it feature also  
y = np.array(df.Salary).reshape(-1, 1) # out put or target variable 

# what the reshape will do is that we have a list of salary [1,2,4,5] for example, it will change it to [ [1,2,4,5] ] - which is 2D array

print(x)
print(y)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=x, y=y, data=df, color="teal")
plt.title("Regression Plot of X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python

x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2, random_state=0)
model = LinearRegression().fit(x_train,y_train )
```

```python
model.predict([[10] ,[4],[3],[11]])
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color='darkblue')
```

```python
# Accuracy of our model
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```
