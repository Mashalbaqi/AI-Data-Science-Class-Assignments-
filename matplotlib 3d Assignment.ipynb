```python
# What is fig?
# In Matplotlib, a Figure is the top-level container that holds everything that will be displayed on the screen or saved to a file. It serves as the canvas or page where all the plots, axes, labels, and other visual elements are placed.

# Think of the Figure as the entire figure windowâ€”the background that can contain multiple plots
```

```python
# What is Axes?
# In Matplotlib, an Axes object represents the area where data is plotted. Each individual plot, graph, or chart is managed by an Axes object. It's one of the most essential elements in Matplotlib, as it controls not just the data but also the ticks, labels, grids, and legends associated with that particular plot.
# The Axes object is the actual plotting area (the drawing surface) in a Figure.
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot([1, 2, 3], [4, 5, 6], [7, 8, 9])
ax.set_xlabel('X ')
ax.set_ylabel('Y ')
ax.set_zlabel('Z ')
plt.show()
```

```python
# 3D Line Plot
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1,2,3,4,5]
y = [10, 20, 30, 40, 50]
z = [100, 200, 300, 400, 500]

ax.plot(x, y, z, label='3D Line')
ax.legend()

plt.show()
```

```python
# 3D Line Plot
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [15,20,30,90,55]
y = [10, 33, 60, 25, 50]
z = [66, 45, 76, 80, 65]

ax.plot(x, y, z, label='3D Line')
ax.scatter(x, y, z, c='r', marker='*')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()

```

```python
# 3D Surface Plot
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X = np.linspace(-3, 3, 30)
Y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X = np.linspace(-3, 3, 30)
Y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**3 + Y**3))

ax.plot_wireframe(X, Y, Z, color='b')

plt.show()
```
